```python
from datetime import datetime
```

```python
start_time = datetime.now()
###number of positive/negative/neutral aspects per POI
num_pos_asp=[]
num_neg_asp=[]
num_neu_asp=[]
for i in range(519):
    review_id_range_start=aspect_sentiment_agg['REVIEW_ID'] >= (sum(num_of_review[0:i]))
    review_id_range_end=aspect_sentiment_agg['REVIEW_ID'] < (sum(num_of_review[0:i+1]))
    current_poi=aspect_sentiment_agg[review_id_range_start & review_id_range_end]
    pos=current_poi[current_poi['SENTIMENT_LABEL']=='positive'].shape[0]
    neg=current_poi[current_poi['SENTIMENT_LABEL']=='negative'].shape[0]
    neu=current_poi[current_poi['SENTIMENT_LABEL']=='neutral'].shape[0]
    num_pos_asp+=[pos]
    num_neg_asp+=[neg]
    num_neu_asp+=[neu]

end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```

```python
start_time = datetime.now()
###number of positive/negative/neutral aspects per POI(non-loop version)
def find_asp_num(poi_index):
    review_id_range_start=aspect_sentiment_agg['REVIEW_ID'] >= (sum(num_of_review[0:poi_index]))
    review_id_range_end=aspect_sentiment_agg['REVIEW_ID'] < (sum(num_of_review[0:poi_index+1]))
    current_poi=aspect_sentiment_agg[review_id_range_start & review_id_range_end]
    pos=current_poi[current_poi['SENTIMENT_LABEL']=='positive'].shape[0]
    neg=current_poi[current_poi['SENTIMENT_LABEL']=='negative'].shape[0]
    neu=current_poi[current_poi['SENTIMENT_LABEL']=='neutral'].shape[0]
    return [pos,neg,neu]
asp_num=list(map(find_asp_num,list(range(519))))
num_pos_asp=[item[0] for item in asp_num]
num_neg_asp=[item[1] for item in asp_num]
num_neu_asp=[item[2] for item in asp_num] 

end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```

```python
###get percentage of sentiment
def find_asp_percentage(poi_index):
    try:
        current_value=asp_num[poi_index]
        total=sum(current_value)
        pos_percentage=current_value[0]/total
        neg_percentage=current_value[1]/total
        neu_percentage=current_value[2]/total
    except:
        pos_percentage=0
        neg_percentage=0
        neu_percentage=0
    return [pos_percentage,neg_percentage,neu_percentage]
asp_percentage=list(map(find_asp_percentage,list(range(519))))
per_pos_asp=[item[0] for item in asp_percentage]
per_neg_asp=[item[1] for item in asp_percentage]
per_neu_asp=[item[2] for item in asp_percentage] 
```

```python
value={'POI_INDEX': list(range(1,519)),
       'NUMBER_OF_REVIEW':num_of_review[1:],
       'NUM_POS_ASP':num_pos_asp[1:],
       'NUM_NEG_ASP':num_neg_asp[1:],
       'NUM_NEU_ASP':num_neu_asp[1:],
       'PER_POS_ASP':per_pos_asp[1:],
       'PER_NEG_ASP':per_neg_asp[1:],
       'PER_NEU_ASP':per_neu_asp[1:]}
attraction_sentiments=pd.DataFrame(data=value)
```

```python
###get the most positively reviewed poi
review_number_filter = attraction_sentiments[attraction_sentiments['NUMBER_OF_REVIEW']>500]
attraction_sentiments_sorted=review_number_filter.sort_values(by='PER_POS_ASP',ascending=False)
attraction_sentiments_sorted.head(5)
```
