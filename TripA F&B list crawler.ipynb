```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd
import regex
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import time
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys ## give access to enter, esc keys etc
from selenium.webdriver.common.by import By
import csv
```

```python
with open('crawl.csv','w',encoding="utf-8") as f:
    f.write("Restaurant_names,Restaurant_type,review_no,url\n")

PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)

restaurant_names =[]
restaurant_review_no = []
restaurant_types =[]
restaurant_urls = []

driver.get("https://www.tripadvisor.com.sg/Restaurants-g294265-Singapore.html")

time.sleep(10)

restaurant_name = driver.find_elements_by_xpath('//a[@class="_15_ydu6b"]')
review_no = driver.find_elements_by_xpath('//span[@class="w726Ki5B"]')
restaurant_type = driver.find_elements_by_xpath('//span[@class="_1p0FLy4t"]')
restaurant_url = driver.find_elements_by_xpath('//a[@class="_15_ydu6b"]')


for name in restaurant_name:
    print(name.text)
    restaurant_names.append(name.text)

for review in review_no:
    print(review.text)
    restaurant_review_no.append(review.text)

for type in restaurant_type:
    print(type.text)
    restaurant_types.append(type.text)

for elem in restaurant_url:
    print(elem.get_attribute("href"))
    restaurant_urls.append(elem.get_attribute("href"))

for item in restaurant_types:
    if "Open" in item:
        restaurant_types.pop(restaurant_types.index(item))
        
for item in restaurant_types:
    if "Close" in item:
        restaurant_types.pop(restaurant_types.index(item))
        
for item in restaurant_types:
    if "reviews" in item:
        restaurant_types.pop(restaurant_types.index(item))
for item in restaurant_types:
    if "$" in item:
        restaurant_types.pop(restaurant_types.index(item))
        
restaurant_types = [x.replace(",", "") for x in restaurant_types]
restaurant_review_no = [x.replace(",", "") for x in restaurant_review_no]

with open('crawl.csv','a') as f:
    for i in range(len(restaurant_name)):
        f.write(restaurant_name[i].text + "," + restaurant_types[i] + "," + restaurant_review_no[i] + "," 
                + restaurant_url[i].get_attribute("href") +"\n")
        
        
driver.execute_script("scroll(0, document.body.scrollHeight);")

while True:
    next_page_btn = driver.find_elements_by_xpath('//*[@id="EATERY_LIST_CONTENTS"]/div[2]/div/a')
    if len(next_page_btn) < 1:
        print("No more pages left")
        break
    else:
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, 'Next'))).click()
        time.sleep(10)
        
        
        restaurant_name = driver.find_elements_by_xpath('//a[@class="_15_ydu6b"]')
        review_no = driver.find_elements_by_xpath('//span[@class="w726Ki5B"]')
        restaurant_type = driver.find_elements_by_xpath('//span[@class="_1p0FLy4t"]')
        restaurant_url = driver.find_elements_by_xpath('//a[@class="_15_ydu6b"]')


        for name in restaurant_name:
            print(name.text)
            restaurant_names.append(name.text)

        for review in review_no:
            print(review.text)
            restaurant_review_no.append(review.text)

        for type in restaurant_type:
            print(type.text)
            restaurant_types.append(type.text)
        for elem in restaurant_url:
            print(elem.get_attribute("href"))
            restaurant_urls.append(elem.get_attribute("href"))

        for item in restaurant_types:
            if "Open" in item:
                restaurant_types.pop(restaurant_types.index(item))
        
        for item in restaurant_types:
            if "Close" in item:
                restaurant_types.pop(restaurant_types.index(item))
        
        for item in restaurant_types:
            if "reviews" in item:
                restaurant_types.pop(restaurant_types.index(item))
        for item in restaurant_types:
            if "$" in item:
                restaurant_types.pop(restaurant_types.index(item))
        
        restaurant_types = [x.replace(",", "") for x in restaurant_types]
        restaurant_review_no = [x.replace(",", "") for x in restaurant_review_no]

        with open('crawl.csv','a',encoding="utf-8") as f:
            for i in range(len(restaurant_name)):
                f.write(restaurant_name[i].text + "," + restaurant_types[i] + "," + restaurant_review_no[i] + "," 
                        + restaurant_url[i].get_attribute("href") +"\n")
driver.quit()  
```
