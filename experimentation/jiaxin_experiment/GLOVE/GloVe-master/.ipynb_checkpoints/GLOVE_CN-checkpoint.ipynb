```python
import numpy as np
import pandas as pd
import random
```

```python
df = pd.read_csv('vectors.txt', delim_whitespace=True, header=None)
df.head()
```

```python
matrix = df.iloc[:, 1:].values
matrix.shape
```

```python
no_of_categories = 11
```

```python
def get_centroid(matrix):
    return np.mean(matrix, axis=0)
```

```python
## mahatten, euclidean, cosine similarity, correlation

def get_manhattan(a, b):
    return np.abs(a - b).sum()

def get_euclidean(a, b):
    return np.linalg.norm(a-b)

def get_cos_sim(a, b):
    return np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))

def get_corr(a, b):
    return np.correlate(a, b)[0]
```

```python
# obtain the index for wrods from each category
index_fac=[44,187,191,269,288,342,348]
index_act=[12,42,85,123,126,136,173]
index_env=[35,54,79,100,147,153,158]
index_exp=[71,101,144,179,211,240,317]
index_fb=[38,68,74,167,262,271,296]
index_fam=[2,9,59,69,87,114,119]
index_pr=[13,24,26,97,110,122,169]
index_ser=[72,138,199,237,508,574,782]
index_sh=[80,159,192,303,304,411,463]
index_ss=[1,6,7,8,14,15,17]
index_trans=[28,60,75,76,84,86,96]
```

```python
fac=matrix[index_fac,:]
act=matrix[index_act,:]
env=matrix[index_env,:]
exp=matrix[index_exp,:]
fb=matrix[index_fb,:]
fam=matrix[index_fam,:]
pr=matrix[index_pr,:]
ser=matrix[index_ser,:]
sh=matrix[index_sh,:]
ss=matrix[index_ss,:]
trans=matrix[index_trans,:]
mtx_cat=[fac,act,env,exp,fb,fam,pr,ser,sh,ss,trans]
```

```python
#dict version, for fun           >_<
#read the list of vectors from the index and calculate centroid
list_cat={'facility':index_fac,
          'activities':index_act,
          'environment':index_env,
          'experiment':index_exp,
          'f&b':index_fb,
          'family':index_fam,
          'price':index_pr,
          'service':index_ser,
          'shopping':index_sh,
          'sightseeing':index_ss,
          'transportaiton':index_trans}
```

```python
matrix_cat={'facility':[],
          'activities':[],
          'environment':[],
          'experiment':[],
          'f&b':[],
          'family':[],
          'price':[],
          'service':[],
          'shopping':[],
          'sightseeing':[],
          'transportaiton':[]}
```

```python
for matkey in matrix_cat:
    init=np.zeros((7,300))
    for j in range(len(list_cat[matkey])):
        init[j]=matrix[list_cat[matkey][j],:]
    matrix_cat[matkey]=init
```

```python
matrix_cat
```

```python
centroids = []
for i in mtx_cat:
    centroids.append(get_centroid(i))
centroids
```

```python
manhattan_distances = np.zeros((matrix.shape[0], no_of_categories))
euclidean_distances = np.zeros((matrix.shape[0], no_of_categories))
cos_sim = np.zeros((matrix.shape[0], no_of_categories))
correlations = np.zeros((matrix.shape[0], no_of_categories))

def get_distances(centroids, matrix):
    for i in range(matrix.shape[0]):
        for j in range(no_of_categories):
            manhattan_distances[i, j] = get_manhattan(matrix[i], centroids[j])
            
    for i in range(matrix.shape[0]):
        for j in range(no_of_categories):
            euclidean_distances[i, j] = get_euclidean(matrix[i], centroids[j])
            
    for i in range(matrix.shape[0]):
        for j in range(no_of_categories):
            cos_sim[i, j] = get_cos_sim(matrix[i], centroids[j])
            
    for i in range(matrix.shape[0]):
        for j in range(no_of_categories):
            correlations[i, j] = get_corr(matrix[i], centroids[j])
```

```python
%%time 
get_distances(centroids, matrix)
```

```python
manhattan_distances.shape
```

```python
def parse_to_df(matrix):
    final_df = pd.DataFrame(data=matrix)
    final_df['KEYWORD'] = df[0]
    cols = ['KEYWORD'] + [i for i in range(no_of_categories)]
    colnames = ['KEYWORD'] + [i for i in range(1, no_of_categories+1)]
    final_df = final_df.loc[:, cols]
    final_df.columns = colnames
    return final_df

manhattan_df = parse_to_df(manhattan_distances)
manhattan_df.head()
```

```python
def get_closest_index(row):
    dictionary = dict(zip(row, list(range(1, no_of_categories+1))))
    return dictionary[min(row)]

manhattan_df['CLOSEST'] = manhattan_df.iloc[:,1:no_of_categories+1].apply(
    get_closest_index, 
    axis=1
)
```

```python
name=['Facilities',
      'Activities',
      'Environment and Ambience',
      'Experimentation',
      'F&B',
      'Family/Childre family/Friends',
      'Price',
      'Service',
      'Shopping',
      'Sightseeing/Place',
      'Transportation and Accessibility']
manhattan_df['NAME']=list(map(lambda x:name[x-1],manhattan_df['CLOSEST']))
manhattan_df.columns
```

```python
manhattan=pd.DataFrame({'KEYWORD':manhattan_df['KEYWORD'],
                               'CAT_INDEX':manhattan_df['CLOSEST'],
                               'CAT_NAME':manhattan_df['NAME']})
```

```python
manhattan.to_csv('./manhattan_distance.csv',index=False)
```

```python
euclidean_df = parse_to_df(euclidean_distances)

euclidean_df['CLOSEST'] = euclidean_df.iloc[:,1:no_of_categories+1].apply(
    get_closest_index, 
    axis=1
)
euclidean_df['NAME']=list(map(lambda x:name[x-1],euclidean_df['CLOSEST']))
euclidean_df
```

```python
euclidean=pd.DataFrame({'KEYWORD':euclidean_df['KEYWORD'],
                               'CAT_INDEX':euclidean_df['CLOSEST'],
                               'CAT_NAME':euclidean_df['NAME']})
```

```python
euclidean.to_csv('./euclidean_distance.csv',index=False)
```

```python
cos_df = parse_to_df(cos_sim)

cos_df['CLOSEST'] = euclidean_df.iloc[:,1:no_of_categories+1].apply(
    get_closest_index, 
    axis=1
)
cos_df['NAME']=list(map(lambda x:name[x-1],cos_df['CLOSEST']))
cos_df
```

```python
cos=pd.DataFrame({'KEYWORD':cos_df['KEYWORD'],
                               'CAT_INDEX':cos_df['CLOSEST'],
                               'CAT_NAME':cos_df['NAME']})
cos.to_csv('./cosine_similarity.csv',index=False)
```

```python
correlations_df = parse_to_df(correlations)

correlations_df['CLOSEST'] =correlations_df.iloc[:,1:no_of_categories+1].apply(
    get_closest_index, 
    axis=1
)
correlations_df['NAME']=list(map(lambda x:name[x-1],correlations_df['CLOSEST']))
correlations_df
```

```python
cor=pd.DataFrame({'KEYWORD':correlations_df['KEYWORD'],
                               'CAT_INDEX':correlations_df['CLOSEST'],
                               'CAT_NAME':correlations_df['NAME']})
cor.to_csv('./correlation.csv',index=False)
```
