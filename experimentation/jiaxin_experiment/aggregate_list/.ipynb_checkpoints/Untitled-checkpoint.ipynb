```python
def parse_reviews(text):
    if re.search(r'\d+',str(text)):
        return int(re.search(r'\d+', text).group())
    else:
        return 0
```

```python
import pandas as pd
import re
```

```python
df = pd.read_csv('20200204_chinese_ctrip_list.csv')
df
```

```python
df['num_reviews'] = df['num_reviews'].apply(lambda x: parse_reviews(x))
```

```python
df = df.dropna()
```

```python
df['link'] = df['link'].apply(lambda x: 'https://you.ctrip.com' + x)
```

```python
df.to_csv('parsed_ctrip_cn_list.csv')
```
