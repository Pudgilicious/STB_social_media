```python
import pandas as pd
```

```python
df1 = pd.read_csv('aspects_200215_074459.csv')
df2 = pd.read_csv('aspects_200215_092433.csv')
df3 = pd.read_csv('aspects_200215_111623.csv')
```

```python
df = df1.append(df2, ignore_index=True).append(df3, ignore_index=True)
df = df.dropna()
df
```

```python
D = dict()

for i in range(len(df)):
    aspect = df.iloc[i, 1]
    if aspect in D:
        D[aspect] += 1
    else:
        D[aspect] = 1

D
```
